```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import glob
import math

files = glob.glob("states*.csv")
df_list = []
for filename in files:
  data = pd.read_csv(filename, index_col=0)
  df_list.append(data)
us_census = pd.concat(df_list)

# Resetting index and gathering info about the dataframe
#print(us_census.columns)
#print(us_census)
us_census.reset_index(inplace=True, drop=True)

#print(us_census)
#print(us_census.info())
#print(len(us_census.State))
#print(us_census.dtypes)
#print(us_census.State.nunique()) 
# at glance, it appears to be all 50 states but we should still check for duplicates...


# Check for duplicates
duplicates = us_census.duplicated()
#print(duplicates.value_counts())
us_census = us_census.drop_duplicates()

# after removing duplicates, we need to reset index again
us_census.reset_index(inplace=True, drop=True)


# Check if any column has missing values
print(us_census.isnull().sum())
# Here we can see that only one column has missing values: us_census['Pacific']
# However, it appears that the us_census['GenderPop'] column will need to be split into two separate male/female columns
# and that might lead to additional missing values
# I am going to hold off on dealing with missing values for now and first address some other column cleaning


# Clean us_census['Income']
split_income = us_census['Income'].str.split('(\$)', expand=True)
us_census['Income'] = pd.to_numeric(split_income[2])


# Clean us_census['GenderPop']
split_genderpop = us_census['GenderPop'].str.split('(\_)', expand=True) #print(split_genderpop)
# Shouldn't I actually search the gender pop column initially to see if there are any decimals to makes sure 
#i don't accidentally lose them when spliting? except why would i? how to search objects in a series for a value
# checking for decimal periods
bool_series = (split_genderpop[2].str.contains('.', regex=False))
for index, value in bool_series.items():
  if (value == True):
    print(f'index = { index }, value = { value }')   
# end checking for decimal periods
us_census['Male'] = split_genderpop[0]
us_census['Female'] = split_genderpop[2]
split_male = us_census['Male'].str.split('M', expand=True) 
#print("split_male")
#print(split_male)
us_census['Male'] = pd.to_numeric(split_male[0])
split_female = us_census['Female'].str.split('F', expand=True)
#print("split_female") 
#print(split_female)
us_census['Female'] = pd.to_numeric(split_female[0])
# I don't understand why us_census['Male'] and us_census['Female'] ended up defaulting to different data types, 
# but for sake of time, I am just going to accept it and move on.  
#print(us_census)
#print(us_census.dtypes)


# Check if any column has missing values
print(us_census.isnull().sum())
# based on this, I know that us_census['Female'] and us_census['Pacific'] have missing values

# Filling missing Values in us_census.Female
#print(us_census.Female.isnull().sum())
us_census = us_census.fillna(value={"Female": (us_census.TotalPop - us_census.Male)})
#print(us_census.Female.isnull().sum())
#print(us_census)


# It makes more sense to address the missing values in us_census['Pacific'] 
# after first cleaning the column by removing the percentage signs and converting it to a number

# remove % from us_census['Pacific']
us_census.Pacific = us_census['Pacific'].replace('[\%,]', '', regex=True)

# make us_census['Pacific'] a number aka put it in numerical format
us_census['Pacific'] = pd.to_numeric(us_census['Pacific'])

# now we can replace the missing values in us_census['Pacific'] with the mean
us_census = us_census.fillna(value={"Pacific": us_census.Pacific.mean()})
# do we care that we lost a few small digits after the decimal point?

#print(us_census.isnull().sum())
# I successfully dealt with the remaining missing values in us_census


# Clean the rest of the race variables: Remove unnecessary symbols, etc.; Convert to number

# remove % from us_census['Hispanic']
us_census.Hispanic = us_census['Hispanic'].replace('[\%,]', '', regex=True)

# make us_census['Hispanic'] a number aka put it in numerical format
us_census['Hispanic'] = pd.to_numeric(us_census['Hispanic'])


# remove % from us_census['White']
us_census.White = us_census['White'].replace('[\%,]', '', regex=True)

# make us_census['White'] a number aka put it in numerical format
us_census['White'] = pd.to_numeric(us_census['White'])


# remove % from us_census['Black']
us_census.Black = us_census['Black'].replace('[\%,]', '', regex=True)

# make us_census['Black'] a number aka put it in numerical format
us_census['Black'] = pd.to_numeric(us_census['Black'])


# remove % from us_census['Native']
us_census.Native = us_census['Native'].replace('[\%,]', '', regex=True)

# make us_census['Native'] a number aka put it in numerical format
us_census['Native'] = pd.to_numeric(us_census['Native'])


# remove % from us_census['Asian']
us_census.Asian = us_census['Asian'].replace('[\%,]', '', regex=True)

# make us_census['Asian'] a number aka put it in numerical format
us_census['Asian'] = pd.to_numeric(us_census['Asian'])

# I successfully removed % signs and to_numeric() for each of the remaining us_census race variables
```
